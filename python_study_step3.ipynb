```python
# 문제 1. Person 클래스 구현
# 요구사항
# 속성: name (이름), age (나이)
# 메서드: introduce()
# 출력 예시:


class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age

person = Person("go", 13)

print(person)

print(person.name)
print(person.age)
```

```python
# 4. 학점 매기기
# 점수 리스트 scores를 받아, 각 점수에 대응하는 학점(A/B/C/D/F)을 튜플 (score, grade) 형태로 모아 반환하는 함수 assign_grades(scores)를 작성하세요.
# 90점 이상: A
# 80점 이상: B
# 70점 이상: C
# 60점 이상: D
# 그 외: F
# 예시



def assign_grades(scores):

  lst = []

  for i in scores:

    if i >= 90:

      grade = 'A'

    if i >= 80:

      grade = 'B'

    if i >= 70:

      grade = 'C'

    if i >= 60:

      grade = 'D'

    if i >= 50:

      grade = 'F'

    lst.append((i, grade))

  return lst



for score, grd in assign_grades([95, 82, 74, 61, 50]):
    print(f"{score}점 → {grd}")
```

```python
# 5. 특정 길이 이상 문자열 필터링
# 문자열 리스트 words와 정수 min_len을 받아, 글자 수가 min_len 이상인 단어만 모아
# 새 리스트로 반환하는 함수 filter_long_strings(words, min_len)를 작성하세요.


def filter_long_strings(words, min_len):

  lst = []

  for i in words:

    if len(i) >= min_len:

      lst.append(i)

    else:
      continue

  return(lst)
fruits = ["apple", "banana", "kiwi", "cherry", "fig"]

print(filter_long_strings(fruits, 4))




# 예시
fruits = ["apple", "banana", "kiwi", "cherry", "fig"]
print(filter_long_strings(fruits, 5))
# 출력: ['apple', 'banana', 'cherry']
print(filter_long_strings(fruits, 4))
# 출력: ['apple', 'kiwi', 'cherry']
```

```python
#Rectangle 클래스 구현
#요구사항

#속성: width (가로 길이), height (세로 길이)

#메서드:

#area() : 면적 반환
#perimeter() : 둘레(=2*(width+height)) 반환


class Rectangle:
  def __init__(self, width, height):

    self.width = width
    self.height = height

  def area(self):

    return self.width * self.height

  def perimeter(self):

    return 2 * (self.width + self.height )


rectangle = Rectangle(9,8)

print(rectangle.area())
print(rectangle.perimeter())
```

```python
# 문제 3. Circle 클래스 구현
# 요구사항
#
# 속성: radius (반지름)
#
# 메서드:
#
# area() : 원의 넓이(πr²) 반환
#
# circumference() : 원의 둘레(2πr) 반환
#
# 힌트: import math 후 math.pi 사용

import math

class Circle:
  def __init__(self, radius):

    self.radius = radius

  def area(self):

    return math.pi * (self.radius ** 2)

  def circumference(self):

    return 2 * math.pi * self.radius

circle = Circle(4)

print(circle.area())

print(circle.circumference())
```

```python
# 문제 4. Student 클래스 구현
# 요구사항
#
# 속성: name (이름), scores (점수 리스트)
#
# 메서드:
#
# average_score() : 평균 점수 반환
#
# print_result() :
#
# 평균이 60 이상이면 "합격", 아니면 "불합격" 출력

class Student:
  def __init__(self, name, scores):

    self.name = name
    self.scores = scores

  def average_score(self):
    return sum(self.scores) / len(self.scores)

  def print_result(self):

    avg = self.average_score()

    if avg >= 60:

      print(f"{self.name} 학생의 평균 점수는 {avg:.1f}점입니다.")

    else:

      print(f"{self.name} 학생의 평균 점수는 {avg:.1f}점입니다.")

s1 = Student("김영희", [80, 75, 90])
s2 = Student("이철수", [50, 40, 60])

s1.print_result()
s2.print_result()
```
